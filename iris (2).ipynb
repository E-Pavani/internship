```python
import pandas as pd
s=pd.read_csv('iris.csv')
s
```

```python
print(s.isnull())#checking the null values
```

```python
print(s.dtypes)#finding the data types 
```

```python
s['species']=s['species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
print(s)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
sns.set(style = "whitegrid")
plt.figure(figsize=(10,6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=s)
plt.title("Scatter plot of Petal length and Petal width")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=s)
plt.title('Scatterplot of Sepal length and sepal width')
plt.show()
```

```python
a=s.drop(columns='species')
a
```

```python
b=s['species']
b
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
at,atest,bt,btest=train_test_split(a,b,test_size=0.2)
model=LogisticRegression()
model.fit(at,bt)
predictions=model.predict(at)
print("b_test shape : ",bt.shape)
print("Predictions shape : ",predictions.shape)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(bt,predictions)
print(accuracy)
```
